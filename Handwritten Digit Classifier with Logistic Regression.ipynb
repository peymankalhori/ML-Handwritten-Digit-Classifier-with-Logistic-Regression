```python
x =digits.data
y = digits.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = linear_model.LogisticRegression()
```

یک ریگریشن غیر خطی که کمک میکنه چندتا ناحیه را از هم جدا کنیم.
LogisticRegression

```python
X_train.shape
#اندازه ورودی های ما 64 بعد است
```

```python
model.fit(X_train, y_train)
```

```python
out = model.predict(X_test)
```

```python
#میخوایم ببینیم چندتا درست حدس زده
er = y_test-out
```

```python
er
```

```python
#چه تعداد اشتباه کرده است
sahih=0
ghalat=0
for a in er:
  if(a==0):
    sahih = sahih +1
  if(a!=0):
    ghalat=ghalat+1

print(sahih)
print(ghalat)
```

```python
from sklearn.metrics import confusion_matrix

# --- بخش اول: ماتریس آشفتگی (برای دیدن اینکه کجاها اشتباه کرده) ---
plt.figure(figsize=(10, 7))
cm = confusion_matrix(y_test, out) # مقایسه واقعیت با پیش‌بینی
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=range(10), yticklabels=range(10))
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix:')
plt.show()

# --- بخش دوم: نمایش تصویری پیش‌بینی‌ها (برای دیدن کارکرد مدل روی عکس) ---
# انتخاب 10 تصویر تصادفی از داده‌های تست
plt.figure(figsize=(15, 5))
for i in range(10):
    idx = np.random.randint(0, len(X_test)) # انتخاب یک ایندکس تصادفی
    
    plt.subplot(2, 5, i + 1)
    plt.imshow(X_test[idx].reshape(8, 8), cmap='gray') # نمایش تصویر
    
    # رنگ سبز اگر درست حدس زده باشد، قرمز اگر غلط باشد
    color = 'green' if out[idx] == y_test[idx] else 'red'
    
    plt.title(f"Real: {y_test[idx]}\nPred: {out[idx]}", color=color)
    plt.axis('off')

plt.tight_layout()
plt.suptitle("Visualizing Predictions: Green=Correct, Red=Wrong", fontsize=16, y=1.05)
plt.show()
```
