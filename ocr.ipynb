```python
x =digits.data
y = digits.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
model = linear_model.LogisticRegression()
```

```python
X_train.shape
#اندازه ورودی های ما 64 بعد است
```

```python
model.fit(X_train, y_train)
```

```python
out = model.predict(X_test)
```
